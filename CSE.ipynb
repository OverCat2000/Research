```python
import json
import pandas as pd
```

```python
# Load JSON data from a file
file_path = 'all.json'  # Replace with your file path

with open(file_path, 'r') as file:
    json_data = json.load(file)

# Extract security company name and sector
securities = json_data['data']['watch']
extracted_data = []

for security in securities:
    company_name = security['companyname']
    sector = security['sector']
    ticker = security['security']
    extracted_data.append({'company_name': company_name, 'security': ticker, 'sector': sector})

# Create a pandas DataFrame
df = pd.DataFrame(extracted_data)
```
